# Reading table by BeautifulSoup
## 1. selecting table

```python
url = 'https://tw.money.yahoo.com/currency-converter'
# url = 'http://rate.bot.com.tw/Pages/Static/UIP003.zh-TW.htm'
res = requests.get(url)
res.encoding = 'utf8'
# tables = pd.read_html(res.text)
soup = BeautifulSoup(res.text)
ts = soup.find_all('table')
print len(ts)
print ts[1].text
table = ts[1]
print type(ts[1])
```

## 2. parsing table and saving into nested list

```python
res_table = []
for tr in table.findAll('tr'):
    tds = tr.findAll('td')
    if len(tds) > 0:
        res_table.append([td.text.strip() for td in tds])
        line = '\t'.join(res_table[-1])
#         print(line)
print ','.join(res_table[0])
# print (len(res_table), len(res_table[0]))
```

## 3. Converting nested list to dataframe

```python
import numpy as np
a = np.array(res_table[1:])

mydf = pd.DataFrame(a[:, :4], columns = ['cash','bank', 'buy_in', 'sold_out'])
# mydf['cash']
mydf.cash
```

# Missing value processing

```python
url = 'http://www.society.taichung.gov.tw/section/index-1.asp?Parser=99,16,257,,,,3807,589,,,,42,,3'
careCenter = pd.read_html(url)[1]
```

```python
careCenter
```

```python
careCenter.columns = careCenter.ix[1]
careCenter = careCenter[2:]
# careCenter.head()
careCenter[u'行政  區域']
```

```python
careCenter.columns
```

## Processing mission value

```python
careCenter[u'機構名稱']
```

```python
careCenter[u'行政  區域'] = (
    careCenter[u'行政  區域'].fillna(method='ffill')
)
careCenter[u'行政  區域']
```

```python
import re
def nursing(t):
    match = re.findall(u'養護(\d+)', t)
    if match:
        return int(match[0])
    else:
        return None
careCenter[u'⺫前收容人數(養護)'] = careCenter[u'目前收容人數'].apply(nursing)
careCenter[u'床型/床數(養護)'] = careCenter[u'床型/床數'].apply(nursing)
careCenter
```
