# ‰æùÁÖß‰∏äËø∞Êü•Ë©¢ÁöÑÁµêÊûúÂª∫Á´ãÁ∂≤Áµ°Ë≥áÊñô

```python
edge_dict = {}
for tweet in tweets:
#     text = ''.join(ch for ch in tweet['text'] if ch not in exclude)
    text = tweet['text'].lower()

#     print text, '[@%s]'%(tweet['user']['name'])
    words = text.split()
    hashtags = []
    for w in words:
        if w.startswith('#'):
            hashtags.append(w)
    if len(hashtags) > 1:
        for t1 in hashtags:
            for t2 in hashtags:
                if t1 != t2:
                    if t1 not in edge_dict:
                        edge_dict[t1] = {}
                    if t2 not in edge_dict:
                        edge_dict[t2] = {}
                    if t2 not in edge_dict[t1]:
                        edge_dict[t1][t2] = 0
                    if t1 not in edge_dict[t2]:
                        edge_dict[t2][t1] = 0
                    edge_dict[t1][t2] += 1
                    edge_dict[t2][t1] += 1
for t1, v in edge_dict.items():
    for t2, k in v.items():
        if k > 2:
            print "%s\t%s\t%s"%(t1, t2, k)
```
